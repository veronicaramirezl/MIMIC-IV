## Introduction <a name='introduction'></a>
MIMIC-IV is a freely available electronic health record (EHR) dataset encompassing a decade of patient information (2008-2019) from Beth Israel Deaconess Medical Center [1]. It surpasses its predecessor, MIMIC-III, with a better structure and additional patient information [2, 3].

The dataset draws upon two primary sources: a comprehensive hospital-wide EHR system and an ICU-specific clinical information system [1]. Rigorous de-identification procedures ensure patient privacy while preserving the data's scientific integrity.

It includes vital signs, diagnoses, medications, procedures, and de-identified clinical notes [1]. Researchers leverage MIMIC-IV to investigate disease progression, train machine learning models, and develop clinical decision support tools [1, 2].

As a result, null dates of death indicate the patient was alive at least up to that time point. Inferences regarding patient death beyond one year cannot be made using MIMIC-IV (as per the PhyioNet website) [2]. The majority of patient death information is acquired from hospital records when the individual dies within the BIDMC or an affiliated institute.

MIMIC-IV also offers a dedicated module: MIMIC-IV ECG. This subset focuses specifically on diagnostic electrocardiograms (ECGs) [2]. It includes approximately 800,000 10-second ECG recordings from nearly 160,000 unique patients. Each ECG utilizes 12 leads and is sampled at 500 Hz [3].

```python
# Define a function to calculate the Spearman correlation and check statistical significance
def calculate_spearman_correlation(data, x_col, y_col, strat):
    """
    Calculate Spearman correlation and check statistical significance.

    Parameters:
    - data: DataFrame containing the variables of interest.
    - x_col: Name of the column for the first variable.
    - y_col: Name of the column for the second variable.
    - strat: A string representing the stratification (e.g., 'males' or 'females').

    Returns:
    - rho: Spearman correlation coefficient.
    - p_value: p-value for the correlation test.
    """
    # Calculate Spearman correlation
    rho, p_value = spearmanr(data[x_col], data[y_col])

    # Print results
    print(f'The Spearman correlation coefficient between {x_col} and {y_col} for {strat} is: {rho:.4f}')
    print(f'The p-value is: {p_value:.4f}')

    # Check statistical significance
    if p_value < 0.05:
        print("The correlation is statistically significant.")
    else:
        print("The correlation is not statistically significant.")
    
    return rho, p_value

# Correlation between 'heart_age_gap' and 1-year mortality for males
calculate_spearman_correlation(mortality_m, 'heart_age_gap', 'event_indicator', 'males')

# Correlation between 'heart_age_gap' and 1-year mortality for females 
calculate_spearman_correlation(mortality_f, 'heart_age_gap', 'event_indicator', 'females')
```

```python
# Now, graph the Kaplan-Meier curves for the mortality data in the gender groups

''' 
The Kaplan-Meier test (also known as the Kaplan-Meier estimator) is a non-parametric 
statistical method used to estimate the survival function from time-to-event data. It provides a 
way to visualize and analyze the probability of an event (such as death, disease recurrence, or
recovery) occurring over time.
'''
# Function to perform Kaplan-Meier analysis and plot for each dataset
def kaplan_meier_analysis(mortality_data, condition_name):
    # Drop patients with missing heart age gap data
    mortality_analysis = mortality_data.dropna(subset=['heart_age_gap'])
    
    # Create Kaplan-Meier fitter objects for each heart age category
    age_groups = mortality_analysis["heart_age_category"].unique()
    kmf_results = {}

    for age_group in age_groups:
        subset = mortality_analysis[mortality_analysis["heart_age_category"] == age_group]

        # Initialize KaplanMeierFitter
        kmf = KaplanMeierFitter()

        # Fit the Kaplan-Meier model
        kmf.fit(durations=subset["time_to_event"], event_observed=subset["event_indicator"], label=f"{condition_name} - {age_group}")

        # Store the fitted model
        kmf_results[age_group] = kmf

        # Calculate restricted mean survival time up to a specific time point (e.g., 365 days)
        t_max = 365  # You can adjust this time point as needed
        restricted_mean = restricted_mean_survival_time(kmf, t=t_max)

        print(f"{condition_name} - {age_group}:")
        print(f"Restricted Mean Survival Time (up to {t_max} days): {restricted_mean}")

    # Plot the Kaplan-Meier curves for this condition
    plt.figure(figsize=(10, 6))
    for age_group, kmf in kmf_results.items():
        kmf.plot(label=age_group)

    plt.xlim(0, 365)
    plt.xlabel("Time (days)")
    plt.ylabel("Survival Probability")
    plt.title(f"Kaplan-Meier Curves by Heart Age Category ({condition_name})")
    plt.grid(True)
    plt.legend()
    plt.show()

# Analyze and plot for females
kaplan_meier_analysis(mortality_f, "Female mortality analysis")

# Analyze and plot for males
kaplan_meier_analysis(mortality_m, "Male mortality analysis")

```

```python
# Cox proportional hazards analysis

def cox_proportional_hazards_analysis(mortality_data, condition_name):
    # The dataset should include 'time_to_event', 'event_indicator', and covariates like 'Advanced', 'Decelerated', or others
    cph = CoxPHFitter()
    
    # Fit the Cox model using relevant columns
    cph.fit(mortality_data, duration_col="time_to_event", event_col="event_indicator", formula='Advanced + Decelerated')

    # Print the summary to see hazard ratios and p-values
    print(f"Cox Proportional Hazards Summary for {condition_name}")
    cph.print_summary()

    # Plot the survival function based on the Cox model
    plt.figure(figsize=(10, 6))
    cph.plot()
    plt.title(f"Cox Proportional Hazards Model for {condition_name}")
    plt.grid(True)
    plt.show()

# Analyze and plot for Females
cox_proportional_hazards_analysis(mortality_f, 'Female Mortality')

# Analyze and plot for Males
cox_proportional_hazards_analysis(mortality_m, 'Male Mortality')
```

```python
# Create a new function to categorize the heart age gap

def categorize_heart_age_gap(row):
    '''
    This function categorizes the heart age gap into three categories:
    
    input: row from a DataFrame
    output: heart age category based on the heart age gap
    '''
    if row['heart_age_gap'] > 8.38:
        return 'Advanced'
    elif row['heart_age_gap'] < -8.38:
        return 'Decelerated'
    else:
        return 'Normal'

# Apply the function to the DataFrame
hosp_f['heart_age_category'] = hosp_f.\
    apply(categorize_heart_age_gap, axis=1)

#Lets calculate the time to event for the patients that died

hosp_f['dod'] = hosp_f['dod'].astype('datetime64[ns]')
hosp_f['admittime'] = hosp_f['admittime']\
    .astype('datetime64[ns]')

hosp_f['time_to_event'] = (hosp_f['dod'] - \
    hosp_f['admittime']).dt.days

# Lets censor the patients that are still alive or that died after 365 days of their last 
# hospitalization

hosp_f['event_indicator'] = np.where(
    hosp_f['dod'].isna() | (hosp_f['time_to_event'] > 365), 
    0,  # Still alive or died after 365 days (censored)
    1   # Died within 365 days
)

# Replace all the NaN values in the 'time_to_event' column with 365
hosp_f['time_to_event'] = hosp_f['time_to_event'].fillna(400)

# Create a new column to categorize the heart age gap for males
def categorize_heart_age_gap(row):
    if row['heart_age_gap'] > 8.38:
        return 'Advanced'
    elif row['heart_age_gap'] < -8.38:
        return 'Decelerated'
    else:
        return 'Normal'

# Apply the categorization function to hosp_m
hosp_m['heart_age_category'] = hosp_m.apply(categorize_heart_age_gap, axis=1)

# Calculate time to event for males
hosp_m['dod'] = hosp_m['dod'].astype('datetime64[ns]')
hosp_m['admittime'] = hosp_m['admittime'].astype('datetime64[ns]')
hosp_m['time_to_event'] = (hosp_m['dod'] - hosp_m['admittime']).dt.days

# Censor patients who are still alive or died after 365 days
hosp_m['event_indicator'] = np.where(
    hosp_m['dod'].isna() | (hosp_m['time_to_event'] > 365), 
    0,  # Still alive or died after 365 days (censored)
    1   # Died within 365 days
)

# Replace NaN values in the 'time_to_event' column with 400 days (censored)
hosp_m['time_to_event'] = hosp_m['time_to_event'].fillna(400)

hosp_m.head()
hosp_f.head()
```

```python
# Lets perform a negative binomial regression to see the correlation between the heart age gap 
# and the outcome (Number of hospitalizations in the next 365 days)
'''
Because the data has 3 categories, we will use dummy variables to represent the heart age categories.
This was already done in the last notebook
'''

# Function to perform negative binomial regression and visualize results for each dataset
def negative_binomial_regression(hosp_data, condition_name):
    # Prepare the independent and dependent variables
    X = hosp_data[['Advanced', 'Decelerated']]
    y = hosp_data['hospitalization_count']

    # Add a constant to the independent variables
    X = sm.add_constant(X)

    # Fit the negative binomial regression model
    model = sm.GLM(y, X, family=sm.families.NegativeBinomial()).fit()

    # Get the exponentiated coefficients (exp(coef))
    exp_coef = np.exp(model.params)

    # Display the model summary and exp(coef)
    print(f"Negative Binomial Regression Summary for {condition_name}")
    print(model.summary())
    print("\nExponentiated Coefficients (exp(coef)):")
    print(exp_coef)

    # Visualize the coefficients
    plt.figure(figsize=(10, 6))
    sns.barplot(x=model.params.index, y=model.params.values, palette='viridis')
    plt.title(f'Negative Binomial Regression Coefficients for {condition_name}')
    plt.xlabel('Coefficient')
    plt.ylabel('Value')
    plt.xticks(rotation=45)
    plt.show()


negative_binomial_regression(hosp_f, 'Female Hospitalization')
negative_binomial_regression(hosp_m, 'Male Hospitalization')
```

1. Gender-Stratified Mortality and Hospitalization:

Mortality:
There is no significant correlation between the heart age gap and mortality in this model. The Kaplan-Meier survival curves illustrate the overall population results, showing a lower 1-year survival probability in the decelerated category compared to the advanced heart age category. The Cox proportional hazards model for mortality indicates that decelerated heart age is statistically significant and positively correlated with mortality for both males and females, with a hazard ratio of 1.17. This may be attributed to the model categorizing patients with less adaptive changes, who are closer to a younger heart. In contrast, this lack of adaptation may disadvantage patients whose hearts are not chronically adapted.

Hospitalization:
Advanced heart age consistently correlates with an increased number of hospitalizations for both males and females, while decelerated heart age is associated with fewer hospitalizations. The expected hospitalization count is approximately 18.5% higher in males and 15% higher in patients with a history of infarction for individuals in the "Advanced" group compared to the Normal group. In the decelerated group, females had a 12.2% lower hospitalization count compared to the baseline group, while males had a 13% lower count.
2. Comorbidity-Stratified Analysis:

    Mortality:
For all stratified outcomes, there was a weak correlation between the heart age gap and mortality, which was statistically significant for all analyses.

Diabetes:
Patients with advanced heart age had a lower restricted mean survival time (213.58 days) compared to those with decelerated heart age (246.67 days). This indicates that they died faster than the normal and decelerated groups, marking a change from the overall population findings. The Cox proportional hazards model showed that advanced heart age is a strong predictor of mortality for diabetic patients (hazard ratio = 1.26), while decelerated heart age appears protective (hazard ratio = 0.90). This suggests that heart age may be a predictor of mortality, but only in patients with comorbid diabetes.

Hypertension:
The only statistically significant result was for advanced heart age in patients with hypertension, which showed a negative correlation to mortality according to the Cox model when compared to normal heart age. This finding is consistent with the overall population results but contrasts with the diabetes findings.

Past Infarction:
The restricted mean survival time was lower for patients with advanced heart age, and the Kaplan-Meier analysis confirmed worse outcomes for these patients. The hazard ratio for advanced heart age in the infarction group was 1.11, indicating a moderate directly proportional effect, consistent with the diabetes results. The decelerated group showed no statistically significant relationship with mortality.

 Hospitalization Analysis:

Negative binomial regression analysis revealed that patients with advanced heart age had a significantly higher number of hospitalizations across all comorbidities. The expected hospitalization count is approximately 32% higher in diabetic patients, 52% higher in patients with hypertension, and 28% higher in females in the "Advanced" group compared to the Normal group, with the highest impact observed in hypertension patients. Decelerated heart age was consistently associated with fewer hospitalizations across all conditions, and in the general population, there was an expected count of 21%, 22%, and 20% less compared to Normal patients for diabetic, hypertensive, and past infarction patients, respectively.

3. General Conclusions:

Advanced heart age is a strong predictor of increased hospitalizations, while decelerated heart age offers a protective effect. These findings are consistent across gender and comorbidity strata, supporting the hypothesis that heart age could serve as a predictor for mortality.

For mortality, heart age showed only a weak correlation, with unexpected advanced age predictions of better mortality outcomes in the general population, except for diabetic and past infarction patients. This may be due to confounding factors that increase mortality independently and may be better reflected as aging in previously validated models. These factors could indicate poor health rather than chronic adaptive changes of the heart in elderly heart failure patients.
